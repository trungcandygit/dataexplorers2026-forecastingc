## 05 — Baseline
Đường cơ sở: dự báo tất cả là nhãn đa số (majority class).

```python
# Baseline accuracy = tỷ lệ nhãn đa số
acc_baseline = max(y_train.mean(), 1 - y_train.mean())
print(f"Baseline Accuracy: {acc_baseline:.2f}")
print(f"(Dự báo tất cả là nhãn '{int(y_train.mean() >= 0.5)}' — nhãn đa số)")
```
