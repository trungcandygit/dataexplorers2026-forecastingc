## 06 — XGBoost Model
Huấn luyện XGBoost Classifier với tham số tốt nhất và đánh giá qua 5-fold Stratified CV.

```python
# Tải tham số tốt nhất
params = pd.read_csv("best_params/best_params_classifier.csv", index_col=0)
n_estimators = int(params.loc["n_estimators"][0])
max_depth    = int(params.loc["max_depth"][0])
lr           = float(params.loc["learning_rate"][0])
subsample    = float(params.loc["subsample"][0])

print("XGBoost best params:")
print(f"  n_estimators  = {n_estimators}")
print(f"  max_depth     = {max_depth}")
print(f"  learning_rate = {lr}")
print(f"  subsample     = {subsample}")
```

```python
# Huấn luyện
clf = XGBClassifier(
    n_estimators  = n_estimators,
    max_depth     = max_depth,
    learning_rate = lr,
    subsample     = subsample,
    eval_metric   = "logloss",
    random_state  = 42,
    verbosity     = 0,
)
clf.fit(X_train, y_train)
print("XGBoost fit xong.")
```

```python
# Cross-Validation AUC
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
auc_scores = cross_val_score(clf, X_scaled, y, cv=cv, scoring="roc_auc")
print(f"AUC-ROC (5-fold CV): {auc_scores.mean():.4f} ± {auc_scores.std():.4f}")
```
