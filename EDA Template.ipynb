# Thư viện, thư mục và dữ liệu

```python
#thư viện
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
#tải dữ liệu
#YYYY-MM-DD
df = pd.read_csv('tnbike_data.csv', parse_dates=['ds'])
df.head()
```

```python
#đổi tên biến
df = df.rename(columns={'revenue': 'y'})
df.head(0)
```

# Phân tích dữ liệu thô

```python
#thống kê mô tả
df.describe()
```

```python
#kiểm tra giá trị thiếu
df.isnull().sum()
```

# Biểu đồ tổng quan

```python
#vẽ chuỗi thời gian
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df.ds, df.y/1e9, marker='o', color='steelblue')
ax.set_title('Doanh số theo tháng')
plt.show();
```

# Phân rã chuỗi thời gian

```python
#seasonal decompose
ts = df.set_index('ds')['y']
result = seasonal_decompose(ts, model='additive', period=3)
result.plot();
```

# Kiểm định tính dừng

```python
from statsmodels.tsa.stattools import adfuller
pvalue = adfuller(ts.dropna())[1]
print(f'P-Value ADF: {pvalue:.4f}')
```

# ACF và PACF

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
plot_acf(ts.diff().dropna(), lags=6, ax=ax1)
plot_pacf(ts.diff().dropna(), lags=5, ax=ax2)
plt.tight_layout()
plt.show();
```
