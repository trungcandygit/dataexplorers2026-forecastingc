## 02 — Wrangle Data
Hàm `wrangle()` tải và chuẩn hóa dữ liệu lịch sử doanh số.

> **Lưu ý dữ liệu:** T1–T3/2025 và T1–T3/2026 là dữ liệu thực. T4–T12/2025 là nội suy tuyến tính.

```python
def wrangle(filepath):
    df = pd.read_csv(filepath, parse_dates=["ds"])
    # Đổi tên cột revenue → y (Prophet yêu cầu cột 'ds' và 'y')
    df = df.rename(columns={"revenue": "y"})
    df["ds"] = pd.to_datetime(df["ds"])
    df = df.sort_values("ds").reset_index(drop=True)
    return df

df = wrangle("data/tnbike_data.csv")
print("df shape:", df.shape)
df.head()
```
