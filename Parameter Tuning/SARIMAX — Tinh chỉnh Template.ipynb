# Thư viện và dữ liệu

```python
# !pip install pmdarima
import pmdarima as pm
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../tnbike_data.csv', parse_dates=['ds'])
df = df.rename(columns={'revenue': 'y'})
training = df.iloc[:-3, :]
training.head()
```

# Kiểm định tính dừng

```python
from statsmodels.tsa.stattools import adfuller
pvalue = adfuller(training['y'].dropna())[1]
print(f'P-Value ADF: {pvalue:.4f}')
```

# Auto ARIMA

```python
model = pm.auto_arima(
    training['y'],
    start_p=0, max_p=3, start_q=0, max_q=3,
    d=1, D=1,
    seasonal=True, m=3,
    stepwise=True,
    suppress_warnings=True,
    information_criterion='aic'
)
print(model.summary())
```

# Xuất tham số

```python
best_params_s = pd.DataFrame({'value': {
    'p': model.order[0], 'd': model.order[1], 'q': model.order[2],
    'P': model.seasonal_order[0], 'D': model.seasonal_order[1], 'Q': model.seasonal_order[2]
}})
best_params_s.to_csv('../Forecasting Product/best_params_sarimax.csv')
best_params_s
```
