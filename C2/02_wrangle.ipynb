## 02 — Wrangle Data
Hàm `wrangle()` tải và làm sạch dữ liệu giao dịch.

> ~2.1% đơn thiếu thông tin sản phẩm (product_name/color/group_code) — loại bỏ khi phân tích SKU.
> ~24–29% đơn thiếu customer_code/province_name — đơn lẻ, không phải đại lý.

```python
def wrangle(filepath):
    df = pd.read_csv(filepath, parse_dates=["order_date"])
    # Thêm cột phái sinh
    df["month"] = df["order_date"].dt.to_period("M")
    df["year"]  = df["order_date"].dt.year
    return df

df_full = wrangle("data/fact_sales_full.csv")
print("df_full shape:", df_full.shape)
df_full.head()
```

```python
# Làm sạch: giữ lại dòng có đủ thông tin sản phẩm
df = df_full.dropna(subset=["product_name", "color", "group_code"]).copy()
print(f"Sau clean: {len(df):,} / {len(df_full):,} dòng ({len(df)/len(df_full)*100:.1f}%)")
```

```python
# Kiểm tra giá trị thiếu
missing = df_full.isnull().mean() * 100
print("Missing %:")
print(missing[missing > 0].round(2))
```
